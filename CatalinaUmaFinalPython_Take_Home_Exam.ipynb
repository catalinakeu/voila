## Introduction

Human immunodeficiency virus (HIV) is a viral infection attacking the body’s immune system. With the progression of the disease, an acquired immunodeficiency syndrome (AIDS), the most advanced stage of the disease, may occur. HIV virus is spread via the body fluids such as blood, breast milk, semen and vaginal fluids. With no approved vaccines against HIV, the only current prevention method is causciousness when it comes to blood transfussion, breastfeedeng and intercourse. In this report we will examine how the status of a country affect HIV/AIDS as evaluated through schooling levels. Our hypothesis is that the developed coutris will have much lower HIV/AIDS levels than the developing ones and that in countries where an increase in Schooling leves is present, the HIV/AIDS levels will drop.

## Data preparation, cleaning and exploration

```python
import pandas as pd
factors = pd.read_csv("https://raw.githubusercontent.com/NHameleers/dtz2025-datasets/master/CountryHealthFactors.csv")
factors.head()
```

The first step we took in the data analysis was deleting the coulmns that were irrelevant to our research question, however, we ran into an issue of some coulmn names being 'non-existing'. We decided to use the .rename method to get rid of the extra spaces in some coulmn names.

```python
factors = factors.rename(columns=str.strip)
```

```python
factors_used = factors.drop(['Life expectancy', 'Adult Mortality', 'infant deaths', 'Alcohol', 'percentage expenditure', 'Hepatitis B', 'Measles', 'BMI', 'under-five deaths', 'Polio', 'Total expenditure', 'Diphtheria', 'GDP', 'Population', 'thinness  1-19 years', 'thinness 5-9 years', 'Income composition of resources', ], axis=1)
factors_used.head()
```

## Data visualisation

In order to visually present our data two graphs were created, one showing the change in HIV/AIDS levels over the years in different countries, and the other one showing Schooling levels over the years in different countries. For an easier comparison of the graphs they were arranged horizontally.

```python
import altair as alt
```

```python
HIV_graph = alt.Chart(final).mark_line().encode(
    x='Year:N',
    y='HIV/AIDS',
    color='Country'
).properties(title='HIV/AID versus Year per Country')
Schooling_graph = alt.Chart(final).mark_line().encode(
    x='Year:N',
    y='Schooling',
    color='Country'
).properties(title='Schooling versus Year per Country')
HIV_graph | Schooling_graph
```

All three developed countries, Australia, New Zealand and Iceland, show a steady, low level of HIV/AIDS throughout the years (0.1) as well as a slightly alterating, but still relatively high level of Schooling. Contrary, two developing countries, Djibouti and Niger, show a decrease in HIV/AIDS levels thoroughout the years as well as an increasing level of Schooling. In South Sudan the level of Schooling up until 2010 was 0.0, during which time the HIV/AIDS levels increased. With an increase of Schooling, the HIV/AIDS levels started dropping but were in 2015 still higher than in 2000.

Additionally, a graph showing the severity of HIV/AIDS in a population in a certain country associated with the level of schooling in said country is shown

```python
brush = alt.selection(type='interval')

points = alt.Chart(final).mark_circle().encode(
    x='Schooling',
    y='Country',
    color=alt.condition(brush, 'HIV/AIDS:Q', alt.value('blue')),
    size='HIV/AIDS'
).add_selection(
    brush
)
points
```

An interactive component was added to the graph above where readers can use the slider to select the year and on a graph a display is shown of the HIV/AID level in each country with the responding level of schooling in said year. This interactive visualization allowes for the tracking of HIV/AIDS schooling relationship in different countries through time

```python
import pandas as pd
import altair as alt
import ipywidgets as widgets
from ipywidgets import interact

final = pd.DataFrame(final)

year_slider = widgets.IntSlider(min=2000, max=2015, step=1, value=2010)
widgets.HBox([year_slider])

brush = alt.selection(type='interval')

def interactive_chart(selected_year):
    interactive_data = final[final['Year'] == selected_year]

    chart = alt.Chart(interactive_data).mark_circle(size=100).encode(
        x='Schooling',
        y='Country',
        color=alt.condition(brush, 'HIV/AIDS:Q', alt.value('blue')),
        size=alt.Size('HIV/AIDS:Q', scale=alt.Scale(range=[55, 150]))
    ).properties(
        title=f'Scatter Plot for Year {selected_year}'
    ).add_params(
        brush
    )

    display(chart)

interact(interactive_chart, selected_year=year_slider);
```
